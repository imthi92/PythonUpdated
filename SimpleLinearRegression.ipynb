```python
import pandas as pd #importing pandas library for reading csv file
```

```python
dataset=pd.read_csv("Salary_Data.csv") #assigning IP, OP datas to a varaible "dataset"
```

```python
dataset #calling function of variable
```

```python
independent=dataset[["YearsExperience"]] #assigning input dataset to a variable "independent" for storing it in memory
```

```python
independent #calling function of variable from memory
```

```python
dependent=dataset[["Salary"]] #assigning output dataset to a variable "dependent" for storing it in memory
```

```python
dependent #calling function of variable from memory
```

```python
from sklearn.model_selection import train_test_split #importing filename import from Function/Class

X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

#X-train = training independent dataset 
#X-test = training test dataset 
#Y-train = training dependent dataset 
#Y-train = training dependent dataset 

#train_test_split = inbuilt split functionality operator
```

```python
from sklearn.linear_model import LinearRegression #importing filename import Class/Function name
regressor=LinearRegression()      #Linear regression class functionality assigned to a variable for memory storage 
regressor.fit(X_train, y_train)   #variable with . dot operator to perform input & output training dataset 
```

```python
LinearRegression() #Calling the functionality
```

```python
weight=regressor.coef_ #weight=variable stored with regressor.coef_, 
weight
```

```python
bais=regressor.intercept_ # variable bais stored with regresor.intercept value
bais
```

```python
y_pred=regressor.predict(X_test) #dependent prediction stores with the value regressor.predict test independent dataset
```

```python
from sklearn.metrics import r2_score # importing sklearn.metrics import class/function name
r_score=r2_score(y_test,y_pred)      # r_score variable stored with values 
```

```python
r_score
```
